```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
flights = pd.read_csv("/Users/vickichen/Desktop/flight/midterm-project/data/flights_without_outlier.csv")
```

```python
list(flights.columns)
```

```python
df_flights = flights[['origin','origin_city_name','flights']]
```

```python
df_flights
```

```python
#Total number of flights by city
total_city_flights = df_flights.groupby('origin_city_name').sum().sort_values('flights',ascending=False).head(20)
total_city_flights
```

```python
# percent total per city
total_city_flights['percent_total'] = total_city_flights['flights']/total_city_flights['flights'].sum()

percent_total=total_city_flights['percent_total']

percent_total
```

```python
# cumulate sum of percentage
total_city_flights['cumsum_percent'] = total_city_flights['percent_total'].cumsum()
```

```python
total_city_flights['cumsum_percent']
```

```python
top50_percent = total_city_flights[total_city_flights['cumsum_percent'] <= 0.5]
```

```python
top50_percent.reset_index(inplace=True)
```

```python
top50_percent
```

```python
top50_percent[['city','States']] = top50_percent['origin_city_name'].str.split(',',expand=True)

city_states = top50_percent[['city','States']]


```
