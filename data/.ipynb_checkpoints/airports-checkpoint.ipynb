### Get all airports from flights data 

```python
df_flights_cleaned = pd.read_csv('flights_cleaned.csv', index_col=False)
```

```python
# number of airports
df_unique_airports = pd.DataFrame(df_flights_cleaned.origin.unique(), columns=['airport'])
df_unique_airports
```

```python
df_unique_dest_airports = pd.DataFrame(df_flights_cleaned.dest.unique(), columns=['airport'])
df_unique_dest_airports
```

```python
df_airport_IATA = df_unique_airports.merge(df_unique_dest_airports, how="outer")
df_airport_IATA
```

```python
airports = pd.read_csv('airports_all_info_data.csv')
df_usa_airports = airports[airports['iso_country'].isin(['US'])] 
df_usa_airports.shape
```

```python
df_usa_airports['type'].unique()
```

```python
types = ['medium_airport', 'large_airport']

df_usa_airports = df_usa_airports[df_usa_airports['type'].isin(types)]
```

```python
df_usa_airports.columns
```

```python
cols_to_drop = ['ident', 'gps_code', 'local_code', 'elevation_ft', 'continent']

df_usa_airports.drop(columns=cols_to_drop, inplace=True)
```

```python
# creating longitude and latitude columns by string splitting coordinates column
df_usa_airports[['Longitude', 'Latitude']] = df_usa_airports.coordinates.str.split(", ", expand = True)
df_usa_airports
```

```python
df_usa_airports.drop(columns=['coordinates'], inplace=True)
```

```python
df_usa_airports.to_csv('airports_usa_info_data.csv', index=False)
```
